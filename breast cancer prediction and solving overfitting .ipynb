```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
bc= load_breast_cancer()
```

```python
bc.feature_names
```

```python
df= pd.DataFrame(bc["data"], columns=bc["feature_names"])
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.describe().transpose()
```

```python
df["target"]= bc["target"]
```

```python
df.head()
```

```python
sns.countplot(x= "target", data= df)
```

```python
df.corr()["target"][:-1].sort_values().plot(kind="bar")
```

```python
sns.heatmap(df.corr())
```

```python
X= df.drop("target", axis=1).values
y= df["target"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=101)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler= MinMaxScaler()
```

```python
X_train= scaler.fit_transform(X_train)
X_test= scaler.transform(X_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
```

```python
X_train.shape
```

```python
model= Sequential()
model.add(Dense(30, activation="relu"))
model.add(Dense(15, activation="relu"))

#last activation function will sigmoid beacuse binary classifiaction model
model.add(Dense(1, activation="sigmoid"))
model.compile(loss="binary_crossentropy", optimizer="adam")
```

```python
model.fit(x=X_train, y=y_train, epochs=400, validation_data=(X_test, y_test))
```

```python
pd.DataFrame(model.history.history).plot()
```

```python
from sklearn.metrics import classification_report

predictions = model.predict(X_test)

# Convert probabilities to binary class labels using thresholding
threshold = 0.5  # You can adjust this threshold if needed
predicted_classes = (predictions > threshold).astype(int)
print(classification_report(y_test, predicted_classes))
```

```python
#In this graph we see that after a time validation loss is incress. So our model has overfitting problem
#to solve this problem....
```

```python
model= Sequential()
model.add(Dense(30, activation="relu"))
model.add(Dense(15, activation="relu"))

#last activation function will sigmoid beacuse binary classifiaction model
model.add(Dense(1, activation="sigmoid"))
model.compile(loss="binary_crossentropy", optimizer="adam")
```

```python
#in keras there is a build in function to solve overfitting 
from tensorflow.keras.callbacks import EarlyStopping
```

```python
early_stop= EarlyStopping(monitor="val_loss", mode= "min", verbose=1, patience= 25)
```

```python
model.fit(x=X_train, y=y_train, epochs=400, validation_data=(X_test, y_test),
         callbacks=[early_stop])
```

```python
pd.DataFrame(model.history.history).plot()
```

```python
predictions = model.predict(X_test)

# Convert probabilities to binary class labels using thresholding
threshold = 0.5  # You can adjust this threshold if needed
predicted_classes = (predictions > threshold).astype(int)
print(classification_report(y_test, predicted_classes))
```

```python
#overfitting solve using Dropout
```

```python
model= Sequential()
model.add(Dense(30, activation="relu"))
model.add(Dropout(.5))
model.add(Dense(15, activation="relu"))
model.add(Dropout(.5))
#last activation function will sigmoid beacuse binary classifiaction model
model.add(Dense(1, activation="sigmoid"))
model.compile(loss="binary_crossentropy", optimizer="adam")
```

```python
model.fit(x=X_train, y=y_train, epochs=400, validation_data=(X_test, y_test),
         callbacks=[early_stop])
```

```python
pd.DataFrame(model.history.history).plot()
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
predictions = model.predict(X_test)

# Convert probabilities to binary class labels using thresholding
threshold = 0.5  # You can adjust this threshold if needed
predicted_classes = (predictions > threshold).astype(int)
print(classification_report(y_test, predicted_classes))
```

```python
print(confusion_matrix(y_test, predicted_classes))
```
